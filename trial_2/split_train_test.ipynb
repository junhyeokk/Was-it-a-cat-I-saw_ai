```python
import os
import glob
import shutil
```

```python
cnt = 0
species_dict = dict()

path = "./images/images/*"
for file in glob.glob(path):
    if file[-4:] != ".jpg":
        continue
    
    cnt += 1
    species_name = "_".join(file.split("_")[:-1]).split("/")[-1]
    
    if species_name in species_dict:
        species_dict[species_name].append(file)
    else:
        species_dict[species_name] = list()

# for k in species_dict.keys():
#     print(species_dict[k][0].split('/')[-1])

for k, v in species_dict.items():
    print(f"{k} : {len(v)}")
print(f"total : {cnt}")
```

```python
for species in species_dict.keys():
    print(species)
print(f"total species : {len(species_dict.keys())}")
```

```python
os.mkdir("./train")
os.mkdir("./val")
```

```python
for species, image_list in species_dict.items():
    os.mkdir(f"./train/{species}")
    os.mkdir(f"./val/{species}")
    
    val_cnt = len(image_list) // 10
    # 10분의 1만 validation 데이터로 사용
    
    for i in range(len(image_list)):
        filename = image_list[i].split('/')[-1]
        
        if i < val_cnt:
            shutil.copyfile(image_list[i], f"./val/{species}/{filename}")
        else:
            shutil.copyfile(image_list[i], f"./train/{species}/{filename}")
```

```python
train_path = "./train/*"
validation_path = "./val/*"

print("---train---")
for folder in glob.glob(train_path):
    species_name = folder.split('/')[-1]
    species_cnt = len(glob.glob(f"{folder}/*"))
    print(f"{species_name} : {species_cnt}")
    
print("---validation---")
for folder in glob.glob(validation_path):
    species_name = folder.split('/')[-1]
    species_cnt = len(glob.glob(f"{folder}/*"))
    print(f"{species_name} : {species_cnt}")
```
